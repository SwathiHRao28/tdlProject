```python
import os
import json
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from utils.preprocessing import get_transforms, Vocabulary
import random

class ImageCaptionDataset(Dataset):
    def __init__(self, data_root, split="train", vocab=None, transform=None, debug=False, freq_threshold=5, max_length=20):
        self.data_root = data_root
        self.split = split
        self.transform = transform
        self.debug = debug
        self.freq_threshold = freq_threshold
        self.max_length = max_length

        # Determine image directory based on the split
        # COCO dataset typically stores images in train2014 and val2014 folders
        if split == "train":
            self.img_dir = os.path.join(data_root, "images", "train2014")
        elif split == "val":
            self.img_dir = os.path.join(data_root, "images", "val2014")
        else:
            # Fallback for other splits or if the naming convention changes
            self.img_dir = os.path.join(data_root, "images", split)
            print(f"Warning: Image directory for split '{split}' not explicitly handled. Using '{self.img_dir}'.")

        # Try both naming conventions for captions: simple (captions_train.json) and COCO (captions_train2014.json)
        simple_caption_path = os.path.join(data_root, "captions", f"captions_{split}.json")
        coco_caption_path = os.path.join(data_root, "captions", f"captions_{split}2014.json")

        if os.path.exists(simple_caption_path):
            self.caption_path = simple_caption_path
        elif os.path.exists(coco_caption_path):
            self.caption_path = coco_caption_path
        else:
            # If neither exists, then _load_data will handle generating dummy data
            self.caption_path = simple_caption_path # Default to the simple name for dummy data generation

        self.data = self._load_data()

        # Take small subset for debug mode
        if self.debug:
            self.data = self.data[:100]

        self.vocab = vocab
        if self.vocab is None and split == "train":
            self.vocab = Vocabulary(self.freq_threshold)
            self.vocab.build_vocabulary([item["caption"] for item in self.data])

    def _load_data(self):
        """Loads data from COCO JSON format.
        If data doesn't exist (e.g. dummy run), generates dummy data.
        """
        if os.path.exists(self.caption_path):
            with open(self.caption_path, "r") as f:
                coco_data = json.load(f)

            # Extract image filenames and map image IDs to filenames
            image_id_to_filename = {img['id']: img['file_name'] for img in coco_data['images']}

            processed_data = []
            for ann in coco_data['annotations']:
                image_id = ann['image_id']
                if image_id in image_id_to_filename:
                    processed_data.append({
                        "image_id": image_id_to_filename[image_id],
                        "caption": ann['caption']
                    })
            return processed_data
        else:
            print(f"Warning: {self.caption_path} not found. Using DUMMY data.")
            # Generate dummy data for testing
            os.makedirs(self.img_dir, exist_ok=True)
            dummy_data = []
            for i in range(100 if self.debug else 500):
                img_name = f"dummy_{i}.jpg"
                img_path = os.path.join(self.img_dir, img_name)
                if not os.path.exists(img_path):
                    # Create a random RGB image
                    Image.new('RGB', (224, 224), color=(random.randint(0,255), random.randint(0,255), random.randint(0,255))).save(img_path)

                dummy_data.append({
                    "image_id": img_name,
                    "caption": f"This is dummy image {i} mostly for testing."
                })
            return dummy_data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        img_name = item["image_id"]
        caption = item["caption"]

        img_path = os.path.join(self.img_dir, img_name)
        image = Image.open(img_path).convert("RGB")

        if self.transform is not None:
            image = self.transform(image)
        else:
            image = get_transforms(is_train=False)(image)

        # Numericalize caption
        numericalized_caption = [self.vocab.stoi[self.vocab.start_token]]
        tokenized_caption = self.vocab.numericalize(caption)

        # Ensure the tokenized caption + start/end tokens do not exceed self.max_length
        max_content_length = self.max_length - 2
        if len(tokenized_caption) > max_content_length:
            tokenized_caption = tokenized_caption[:max_content_length]

        numericalized_caption += tokenized_caption
        numericalized_caption.append(self.vocab.stoi[self.vocab.end_token])

        return image, torch.tensor(numericalized_caption), caption


class MyCollate:
    def __init__(self, pad_idx):
        self.pad_idx = pad_idx

    def __call__(self, batch):
        imgs = [item[0].unsqueeze(0) for item in batch]
        imgs = torch.cat(imgs, dim=0)

        targets = [item[1] for item in batch]
        captions = [item[2] for item in batch]

        # Pad targets
        from torch.nn.utils.rnn import pad_sequence
        targets = pad_sequence(targets, batch_first=True, padding_value=self.pad_idx)

        return imgs, targets, captions


def get_loaders(config):
    # Transforms
    train_transform = get_transforms(is_train=True, image_size=config['image_size'])
    val_transform = get_transforms(is_train=False, image_size=config['image_size'])

    # Create datasets
    train_dataset = ImageCaptionDataset(
        data_root=config['data_dir'],
        split="train",
        transform=train_transform,
        debug=config['debug'],
        freq_threshold=config['min_word_freq'],
        max_length=config['max_length']
    )

    # Use train_dataset's vocab for val_dataset
    val_dataset = ImageCaptionDataset(
        data_root=config['data_dir'],
        split="val",
        vocab=train_dataset.vocab, # Use the vocab built from training data
        transform=val_transform,
        debug=config['debug'],
        freq_threshold=config['min_word_freq'], # freq_threshold is not used for val_dataset, but keep for consistency
        max_length=config['max_length']
    )

    # Create collate function
    pad_idx = train_dataset.vocab.stoi["<PAD>"]
    collate_fn = MyCollate(pad_idx=pad_idx)

    # Create DataLoaders
    train_loader = DataLoader(
        dataset=train_dataset,
        batch_size=config['batch_size'] if not config['debug'] else config['debug_batch_size'],
        num_workers=config['num_workers'],
        shuffle=True,
        collate_fn=collate_fn
    )

    val_loader = DataLoader(
        dataset=val_dataset,
        batch_size=config['batch_size'] if not config['debug'] else config['debug_batch_size'],
        num_workers=config['num_workers'],
        shuffle=False, # No need to shuffle validation data
        collate_fn=collate_fn
    )

    return train_loader, val_loader, train_dataset.vocab
```

```python
import os

# Content for utils/preprocessing.py - defines get_transforms and Vocabulary
preprocessing_content = '''
import torch
import torchvision.transforms as transforms
import nltk
from collections import Counter

# Assumed standard image transformation function
def get_transforms(is_train=True, image_size=224):
    if is_train:
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.RandomCrop(image_size, padding=4),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
        ])
    else:
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
        ])

class Vocabulary:
    def __init__(self, freq_threshold):
        self.itos = {0: "<PAD>", 1: "<SOS>", 2: "<EOS>", 3: "<UNK>"}
        self.stoi = {"<PAD>": 0, "<SOS>": 1, "<EOS>": 2, "<UNK>": 3}
        self.freq_threshold = freq_threshold
        self.start_token = "<SOS>"
        self.end_token = "<EOS>"
        self.pad_token = "<PAD>"
        self.unknown_token = "<UNK>"

    def __len__(self):
        return len(self.itos)

    def build_vocabulary(self, sentence_list):
        if not nltk.data.find('tokenizers/punkt'):
            nltk.download('punkt')
        counts = Counter()
        for sentence in sentence_list:
            counts.update(self.tokenizer_eng(sentence))
        for word, count in counts.items():
            if count >= self.freq_threshold:
                self.stoi[word] = len(self.itos)
                self.itos[len(self.itos)] = word

    def tokenizer_eng(self, text):
        return [tok.lower() for tok in nltk.word_tokenize(text)]

    def numericalize(self, text):
        tokenized_text = self.tokenizer_eng(text)
        return [self.stoi[token] if token in self.stoi else self.stoi[self.unknown_token] for token in tokenized_text]
'''

preprocessing_path = 'utils/preprocessing.py'
with open(preprocessing_path, 'w') as f:
    f.write(preprocessing_content)

print(f"Successfully updated {preprocessing_path}.")
```

```python
import os
import torch
import torch.nn as nn
from explainability.attribution import compute_batch_attribution
from explainability.alignment_loss import AlignmentLoss
from explainability.counterfactual import CounterfactualLoss
from torch.utils.tensorboard import SummaryWriter

def train_model(model, train_loader, val_loader, vocab, config, device, max_train_steps=None, max_val_steps=None):
    print(f"[DEBUG] max_train_steps received: {max_train_steps}")
    # Setup
    pad_idx = vocab.stoi[vocab.pad_token]
    criterion_caption = nn.CrossEntropyLoss(ignore_index=pad_idx)
    criterion_align = AlignmentLoss()
    criterion_cf = CounterfactualLoss(mask_ratio=0.2)

    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=config["learning_rate"],
        weight_decay=config["weight_decay"]
    )

    # We will only train decoder and vis_project components
    model.encoder.eval()

    writer = SummaryWriter(log_dir=os.path.join("outputs", "logs"))

    epochs = config["debug_epochs"] if config["debug"] else config["epochs"]
    global_step = 0

    lambda_align = config["alignment_weight"]
    lambda_cf = config["counterfactual_weight"]

    for epoch in range(1, epochs + 1):
        model.train()
        total_loss = 0
        total_caption_loss = 0
        total_align_loss = 0
        total_cf_loss = 0

        for batch_idx, (images, captions, raw_captions) in enumerate(train_loader):
            if max_train_steps is not None and batch_idx >= max_train_steps:
                print(f"Reached max_train_steps ({max_train_steps}) for epoch {epoch}.")
                break

            images = images.to(device)
            captions = captions.to(device)

            # 1. Forward pass for captioning
            # For teacher forcing: input is captions[:, :-1], target is captions[:, 1:]
            targets = captions[:, 1:]
            padding_mask_targets = (targets == pad_idx) # Padding mask for loss calculation

            outputs, attn_weights = model(images, captions, pad_idx)

            # outputs shape: (B, seq_len - 1, vocab_size)
            # targets shape: (B, seq_len - 1)
            caption_loss = criterion_caption(
                outputs.reshape(-1, outputs.shape[-1]),
                targets.reshape(-1)
            )

            # 2. Extract features manually to compute attribution without running encoder again
            with torch.no_grad():
                features = model.encoder(images)
                features = model.vis_project(features)

            loss = caption_loss
            align_loss = torch.tensor(0.0).to(device)
            cf_loss = torch.tensor(0.0).to(device)

            if config["use_alignment_loss"] or config["use_counterfactual_loss"]:
                # 3. Compute Attribution Maps
                # We need input features to track gradients
                features = features.detach().clone()
                features.requires_grad_(True)

                # The actual inputs to decoder for attribution and CF loss
                decoder_input = captions[:, :-1].long() # Ensure text input is Long type
                padding_mask_decoder_input = (decoder_input == pad_idx) # Padding mask for decoder input

                attribution_maps = compute_batch_attribution(model, features, decoder_input, pad_idx, fast_mode=config["debug"])

                if config["use_alignment_loss"]:
                    # Ensure attn_weights shape matches attribution maps (B, seq_len, num_pixels)
                    if attn_weights is not None:
                        # The padding_mask for alignment loss should correspond to the decoder_input sequence length
                        align_loss = criterion_align(attn_weights, attribution_maps, padding_mask_decoder_input)
                        loss = loss + lambda_align * align_loss

                if config["use_counterfactual_loss"]:
                    # Counterfactual logic
                    cf_loss = criterion_cf(model, features, decoder_input, pad_idx, attribution_maps, padding_mask_decoder_input)
                    loss = loss + lambda_cf * cf_loss

            # 4. Backward & Optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            total_caption_loss += caption_loss.item()
            total_align_loss += align_loss.item()
            total_cf_loss += cf_loss.item() if isinstance(cf_loss, torch.Tensor) else cf_loss

            global_step += 1

            if batch_idx % 10 == 0:
                avg_caption_loss = total_caption_loss / (batch_idx + 1)
                avg_align_loss = total_align_loss / (batch_idx + 1)
                avg_cf_loss = total_cf_loss / (batch_idx + 1)

                print(f"Epoch [{epoch}/{epochs}], Step [{batch_idx}], Loss: {total_loss / (batch_idx + 1):.4f} "
                      f"(Cap: {avg_caption_loss:.4f}, Align: {avg_align_loss:.4f}, CF: {avg_cf_loss:.4f})")

                writer.add_scalar('Loss/train', total_loss / (batch_idx + 1), global_step)
                writer.add_scalar('Loss/train_caption', avg_caption_loss, global_step)
                writer.add_scalar('Loss/train_alignment', avg_align_loss, global_step)
                writer.add_scalar('Loss/train_counterfactual', avg_cf_loss, global_step)

        # Validation after each epoch
        model.eval()
        val_total_loss = 0
        val_total_caption_loss = 0
        val_total_align_loss = 0
        val_total_cf_loss = 0

        with torch.no_grad():
            for batch_idx, (images, captions, raw_captions) in enumerate(val_loader):
                if max_val_steps is not None and batch_idx >= max_val_steps:
                    print(f"Reached max_val_steps ({max_val_steps}) for epoch {epoch}.")
                    break

                images = images.to(device)
                captions = captions.to(device)

                targets = captions[:, 1:]
                padding_mask_targets = (targets == pad_idx)

                outputs, attn_weights = model(images, captions, pad_idx)
                caption_loss = criterion_caption(
                    outputs.reshape(-1, outputs.shape[-1]),
                    targets.reshape(-1)
                )

                # Re-extract features for attribution for validation, similar to training
                features = model.encoder(images)
                features = model.vis_project(features)

                loss = caption_loss
                align_loss = torch.tensor(0.0).to(device)
                cf_loss = torch.tensor(0.0).to(device)

                if config["use_alignment_loss"] or config["use_counterfactual_loss"]:
                    decoder_input = captions[:, :-1].long() # Ensure text input is Long type
                    padding_mask_decoder_input = (decoder_input == pad_idx)
                    attribution_maps = compute_batch_attribution(model, features, decoder_input, pad_idx, fast_mode=True)

                    if config["use_alignment_loss"]:
                        if attn_weights is not None:
                            align_loss = criterion_align(attn_weights, attribution_maps, padding_mask_decoder_input)
                            loss = loss + lambda_align * align_loss

                    if config["use_counterfactual_loss"]:
                        cf_loss = criterion_cf(model, features, decoder_input, pad_idx, attribution_maps, padding_mask_decoder_input)
                        loss = loss + lambda_cf * cf_loss

                val_total_loss += loss.item()
                val_total_caption_loss += caption_loss.item()
                val_total_align_loss += align_loss.item()
                val_total_cf_loss += cf_loss.item() if isinstance(cf_loss, torch.Tensor) else cf_loss

            avg_val_loss = val_total_loss / len(val_loader)
            avg_val_caption_loss = val_total_caption_loss / len(val_loader)
            avg_val_align_loss = val_total_align_loss / len(val_loader)
            avg_val_cf_loss = val_total_cf_loss / len(val_loader)

            print(f"Epoch [{epoch}/{epochs}], Validation Loss: {avg_val_loss:.4f} "
                  f"(Cap: {avg_val_caption_loss:.4f}, Align: {avg_val_align_loss:.4f}, CF: {avg_val_cf_loss:.4f})")

            writer.add_scalar('Loss/val', avg_val_loss, global_step)
            writer.add_scalar('Loss/val_caption', avg_val_caption_loss, global_step)
            writer.add_scalar('Loss/val_alignment', avg_val_align_loss, global_step)
            writer.add_scalar('Loss/val_counterfactual', avg_val_cf_loss, global_step)

        # Save model checkpoint
        if epoch % config["save_every"] == 0:
            os.makedirs(config["checkpoint_dir"], exist_ok=True)
            checkpoint_path = os.path.join(config["checkpoint_dir"], f"epoch_{epoch:02d}.pt")
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': total_loss / len(train_loader),
            }, checkpoint_path)
            print(f"Model checkpoint saved to {checkpoint_path}")

    writer.close()
```

```python
print(f"Maximum training steps per epoch: {config['max_train_steps_per_epoch']}")
print(f"Maximum validation steps per epoch: {config['max_val_steps_per_epoch']}")
```
